```python
import pandas as pd

data = pd.read_csv("../data/water_potability.csv", names = ['id', 'pH', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic Carbon', 'Trihalomethanes', 'Turbidity', 'Potability'])
df = pd.DataFrame(data)
```

```python
# Description
desc = df.describe()

# Median
desc.loc['median'] = df.median()

# Variance
desc.loc['variance'] = df.var()

# Range
desc.loc['range'] = df.max() - df.min()

# IQR
desc.loc['IQR'] = df.quantile(0.75) - df.quantile(0.25)

# Skewness
desc.loc['skewness'] = df.skew()

# Kurtosis
desc.loc['kurtosis'] = df.kurtosis()

desc
```
