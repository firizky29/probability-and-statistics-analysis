```python
import pandas as pd

data = pd.read_csv("../data/water_potability.csv", names = ['id', 'pH', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic Carbon', 'Trihalomethanes', 'Turbidity', 'Potability'])
df = pd.DataFrame(data)

data.describe()
```

```python
ph_mean = df['pH'].mean()
ph_median = df['pH'].median()
ph_mode = df['pH'].mode()
ph_std = df['pH'].std()
ph_var = df['pH'].var()
ph_max = df['pH'].max()
ph_min = df['pH'].min()
ph_interval_range = ph_max - ph_min
ph_q1 = df['pH'].quantile([0.25]).values[0]
ph_q3 = df['pH'].quantile([0.75]).values[0]
ph_iqr = ph_q3 - ph_q1
ph_skew = df['pH'].skew()
ph_kurtosis = df['pH'].kurtosis()

print("Untuk data pH")
print("Mean : ", ph_mean)
print("Median : ", ph_median)
#print("Modus : " , ph_mode)
print("Standar Deviasi : ", ph_std)
print("Variansi : ", ph_var)
print("Range : ", ph_interval_range)
print("Quartil 1: ", ph_q1)
print("Quartil 3: ", ph_q3)
print("IQR : ", ph_iqr)
print("Skewness: ", ph_skew)
print("Kurtosis: ", ph_kurtosis)

#boleh pake describe() aja gak sih
```
